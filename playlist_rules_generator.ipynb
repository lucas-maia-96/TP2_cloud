```python
import pandas as pd 
from fpgrowth_py import fpgrowth
```

```python
playlist_basket = []

for chunk in pd.read_csv('2023_spotify_ds1.csv', chunksize=50000):
    basket = chunk.groupby('pid')['track_name'].apply(list).tolist()
    playlist_basket.extend(basket)
```

```python
playlist_basket
```

```python
freqItemSet, rules = fpgrowth(playlist_basket, minSupRatio=0.05, minConf=0.2)
```
